```python
s1 = list(dots1['sum'])
i1 = list(dots1.index)

s2 = list(dots2['sum'])
i2 = list(dots2.index)

s3 = list(dots3['sum'])
i3 = list(dots3.index)

import matplotlib.pyplot as plt
import numpy as np

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,5))
fig.suptitle('Amount of purchases per user')

ax1.scatter(i1, s1,c='g', alpha=0.5, label='Cohort A')
ax1.set_ylim([0, 50])
ax1.axhline(y=40, color='g', linestyle='--')
ax1.grid(True)
ax1.set_xlabel('Users')
ax1.set_ylabel('Toner purchases')
ax1.legend()
ax1.set_title(r'Users = %.f' %(len(s1)))

ax2.scatter(i2, s2,c='b', alpha=0.5, label='Cohort B')
ax2.set_ylim([0, 50])
ax2.grid(True)
ax2.set_xlabel('Users')
ax2.set_ylabel('Toner purchases')
ax2.legend()
ax2.set_title(r'Users = %.f' %(len(s2)))

ax3.scatter(i3, s3,c='r', alpha=0.5, label='Cohort C')
ax3.set_ylim([0, 50])
ax3.axhline(y=30, color='r', linestyle='--')
ax3.grid(True)
ax3.set_xlabel('Users')
ax3.set_ylabel('Toner purchases')
ax3.legend()
ax3.set_title(r'Users = %.f' %(len(s3)))
```

```python
def function(w1):
    for i in range(len(w1)):
        if(i!=0):
            w1[i] = w1[i] + w1[i-1]      
    return w1
```

Amount of items provided by user per cohort

```python
col_list = list(data1)
col_list.remove('user_id')
col_list.remove('cohort')
data1['user_data'] = data1[col_list].sum(axis=1)*100/4.0
```

```python
data1.loc[data1['cohort']=='a','cohort'] = 1
data1.loc[data1['cohort']=='b','cohort'] = 2
data1.loc[data1['cohort']=='c','cohort'] = 3

dots1 = data1.loc[data1['cohort']==1,:]
dots2 = data1.loc[data1['cohort']==2,:]
dots3 = data1.loc[data1['cohort']==3,:]
```

```python
s1 = list(dots1['user_data'])
i1 = list(dots1.index)

s2 = list(dots2['user_data'])
i2 = list(dots2.index)

s3 = list(dots3['user_data'])
i3 = list(dots3.index)

b1 = len(np.unique(s1))
b2 = len(np.unique(s2))
b3 = len(np.unique(s3))

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,5))
fig.suptitle('Frequency of the amount of items in the user data')

(mu, sigma) = norm.fit(s1)

ax1.hist(s1, b1, facecolor='green', alpha=0.75)
ax1.set_title('Cohort A: ' + r'$\mu=%.3f,\ \sigma=%.3f$' %(mu, sigma))
ax1.grid(True)
ax1.set_xlabel('Items of user data percentage')
ax1.set_ylabel('Frequency')

(mu, sigma) = norm.fit(s2)

ax2.hist(s2, b2, facecolor='blue', alpha=0.75)
ax2.grid(True)
ax2.set_xlabel('Items of user data percentage')
ax2.set_ylabel('Frequency')
ax2.set_title('Cohort B: ' + r'$\mu=%.3f,\ \sigma=%.3f$' %(mu, sigma))

(mu, sigma) = norm.fit(s3)

ax3.hist(s3, b3, facecolor='red', alpha=0.75)
ax3.grid(True)
ax3.set_xlabel('Items of user data percentage')
ax3.set_ylabel('Frequency')
ax3.set_title('Cohort C: ' + r'$\mu=%.3f,\ \sigma=%.3f$' %(mu, sigma))
```

Toner purchases by week for Cohort B

```python
dots2 = data2.loc[data2['cohort']==2,:]
i2 = list(dots2.index)

fig, (ax1, ax2, ax3) = plt.subplots(1, 3, figsize=(20,5))
fig.suptitle('Acumulative purchases per week for cohort B')
for i in range(1,16):
    if(i<6):
        ax1.plot(i2, function(list(dots2['week '+ str(i)])), label='week ' + str(i))
        ax1.set_xlabel('Users')
        ax1.set_ylabel('Acumulative purchases')
        ax1.legend()
    elif(i<11 and i > 5):
        ax2.plot(i2, function(list(dots2['week '+ str(i)])), label='week ' + str(i))
        ax2.set_xlabel('Users')
        ax2.set_ylabel('Acumulative purchases')
        ax2.legend()
    else:
        ax3.plot(i2, function(list(dots2['week '+ str(i)])), label='week ' + str(i))
        ax3.set_xlabel('Users')
        ax3.set_ylabel('Acumulative purchases')
        ax3.legend()
```
